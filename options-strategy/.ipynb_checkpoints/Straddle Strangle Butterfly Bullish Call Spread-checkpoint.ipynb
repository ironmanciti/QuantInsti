# Straddle

동일행사가, 동일만기의 put, call option 모두를 premium 지불하고 long position 취함.

주가가 strike price 보다 많이 변동하면 이익. 변하지 않으면 손실

```python
## A 종목을 기초자산으로하는 strike price (ATM) 콜옵션 매수
strike_price_long_call = 1325
premium_long_call = 70
long_call_contracts = 1

# A 종목을 기초자산으로하는 동일한 strike price (ATM) 풋옵션 매수
strike_price_long_put = 1325
premium_long_put = 55
long_put_contracts = 1

# 만기일 주가
sT_expiry = [np.int(strike_price_long_call * 1.1), np.int(strike_price_long_call * 1.05), strike_price_long_call, 
             strike_price_long_put, np.int(strike_price_long_put * 0.95), np.int(strike_price_long_put * 0.9)]

print('Straddle 손익\n')
print('콜옵션 매수 :', strike_price_long_call)
print('풋옵션 매수 :', strike_price_long_put)
```

```python
# Stock price range at expiration of the call
sT = np.arange(0.8*strike_price_long_put,1.2*strike_price_long_put,1) 

payoff_long_call = long_call_payoff(sT, strike_price_long_call, premium_long_call)
payoff_long_put = long_put_payoff(sT, strike_price_long_put, premium_long_put)
payoff_straddle = payoff_long_call + payoff_long_put

fig, ax = plt.subplots()
ax.spines['bottom'].set_position('zero')
ax.plot(sT,payoff_long_call, 'r--', 
        label="long "+str(strike_price_long_call)+" strike call")
ax.plot(sT,payoff_long_put, 'g--', 
        label="long "+str(strike_price_long_put)+" strike put")
ax.plot(sT, payoff_straddle , 'b-', label='Straddle Payoff')
plt.legend()
plt.grid(True)
plt.xlabel('stock A')
plt.ylabel('P & L')
plt.title('Straddle');
```

# Strangle

동일기초자산, 동일 만기의 서로 다른 행가가격의 call 과 put 을 position 으로 보유

## Long Strangle

```python
# A 종목을 기초자산으로 하는 call option 매수
strike_price_long_call = 1250
premium_long_call = 35

# A 종목을 기초자산으로하는 동일  만기의 풋옵션 매수
strike_price_long_put = 1150
premium_long_put = 25

print('Long Strangle 손익\n')
print('콜옵션 매수 :', strike_price_long_call)
print('풋옵션 매수 :', strike_price_long_put)
```

```python
# Stock price range at expiration of the call
sT = np.arange(0.9*strike_price_long_put,1.1*strike_price_long_call,1) 

payoff_long_call = long_call_payoff(sT, strike_price_long_call, premium_long_call)
payoff_long_put = long_put_payoff(sT, strike_price_long_put, premium_long_put)
payoff_straddle = payoff_long_call + payoff_long_put

fig, ax = plt.subplots()
ax.spines['bottom'].set_position('zero')
ax.plot(sT,payoff_long_call, 'r--', 
        label="long "+str(strike_price_long_call)+" strike call")
ax.plot(sT,payoff_long_put, 'g--', 
        label="long "+str(strike_price_long_put)+" strike put")
ax.plot(sT, payoff_straddle , 'b-', label='Straddle Payoff')
plt.legend()
plt.grid(True)
plt.xlabel('stock A')
plt.ylabel('P & L')
plt.title('Strangle long');
```

# Butterfly Spread

4 개의 option 계약으로 구성됨. 동일한 만기일의 3 개의 다른 행사가로 구성.

2 개의 콜옵션 계약을 중간 행사가로 매수. 1 개의 콜옵션은 낮은 행사가로 매도. 1 개의 콜옵션은 높은 행사가로 매도.

Put 과 Call 모두 사용할 수 있음. 이 전략은 기초자산의 가격이 현재 가격에서 많이 변하지 않을 것이라 예상할 때 실행함.

```python
# A 종목을 기초자산으로 하는 call option 2 계약 매수
strike_price_long_call = 1450
premium_long_call = 30

# A 종목을 기초자산으로하는 동일  만기, 낮은 strike 의 콜옵션 1 계약 매도
strike_price_short_call_1 = 1400
premium_short_call_1 = 45

# A 종목을 기초자산으로하는 동일  만기, 높은 strike 의 콜옵션 1 계약 매도
strike_price_short_call_2 = 1500
premium_short_call_2 = 20

total_premium_recieved = - premium_long_call * 2 + premium_short_call_1 + premium_short_call_2

print('Butterfly spread 손익\n')
print('콜옵션 2 계약 매수 :', strike_price_long_call)
print('콜옵션 매도 :', strike_price_short_call_1)
print('콜옵션 매도 :', strike_price_short_call_2)
print()
print('Total premium received : ', total_premium_recieved)
```

```python
# Stock price range at expiration of the call
sT = np.arange(0.98 * strike_price_short_call_1, 1.02 * strike_price_short_call_2, 1) 

payoff_long_call = long_call_payoff(sT, strike_price_long_call, premium_long_call, contracts=2)
payoff_short_call_1 = short_call_payoff(sT, strike_price_short_call_1, premium_short_call_1)
payoff_short_call_2 = short_call_payoff(sT, strike_price_short_call_2, premium_short_call_2)
payoff_butterfly = payoff_long_call + payoff_short_call_1 + payoff_short_call_2

fig, ax = plt.subplots(figsize=(5,6))
ax.spines['bottom'].set_position('zero')
ax.plot(sT,payoff_long_call, 'r--', 
        label="short "+str(strike_price_long_call)+" 2 계약 매도")
ax.plot(sT,payoff_short_call_1, 'g--', 
        label="long "+str(strike_price_short_call_1)+" 1 계약 매수")
ax.plot(sT,payoff_short_call_2, 'k--', 
        label="long "+str(strike_price_short_call_2)+" 1 계약 매수")
ax.plot(sT, payoff_butterfly , 'b-', label='Butterfly Payoff')
plt.legend()
plt.grid(True)
plt.xlabel('stock A')
plt.ylabel('P & L')
plt.title('Butterfly by Call Option');
```

# bull call spread

강세장 예상하나 확신이 서지 않는 경우 사용

행사가격이 높은 콜옵션 매도

그보다 행사가격이 낮은 콜옵션 매입

```python
st = np.arange(900, 1200, 1)
total_payoff = np.zeros(st.shape)
fig, ax = plt.subplots()

# 콜옵션 매도
strike1 = 1075 
premium= 10

payoff = short_call_payoff(st, strike1, premium)
plot_spread(st, payoff, strike1, fig, ax, 'Strike {} call_short'.format(strike1),'dashed')
total_payoff += payoff

# 더 낮은 행사가의 콜옵션 매수
strike2 = 1025
premium= 30

payoff = long_call_payoff(st, strike2, premium)
plot_spread(st, payoff, strike2, fig, ax, 'Strike {} call_long'.format(strike2),'dashed')
total_payoff += payoff

plot_spread(st, total_payoff, strike2, fig, ax, 'Net Spread','solid')

plt.title('Bull Call Spread');
```
