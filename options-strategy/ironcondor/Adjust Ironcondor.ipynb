## When To Adjust Your Iron Condor

## 방법1 : call side 가 위협받을 경우, call side position 을 close 하고 far OTM 으로 reopen

This position was open on September 8th, and looks like this

Bought	 10 OCT 175 PUT	 @ $2.55 (-) 

Sold	 10 OCT 180 PUT	 @ $3.80 (+) 

Sold	 10 OCT 192.5 CALL	 @ $5.50 (+)

Bought	 10 OCT 197.5 CALL	 @ $3.60 (-)

```python
import os
import sys
sys.path.append(os.path.dirname(os.path.abspath(os.path.dirname('./'))))

from payoff import *
```

```python
ironCondor((175, 2.55), (180, 3.80), (192.5, 5.50), (197.5, 3.60),  (170, 205, 1))
```

이제 위 stock 의 가격이 193.92 로 상승하였다고 가정하여 position 을 adjust 한다. 이 가격 상승에 의해 call option 쪽이 영향을 받는다.

Sold      10 OCT 192.5 CALL @ $5.50 (+)

Bought 10 OCT 197.5 CALL @ $3.60 (-)

                                         __________
                                          +$1.90 (Initial credit we had received)
                                          
condor 의 날개는 원래 $\$1.90 $의 credit 으로 open 되었으나 (10 계약 $\$1900$) 193.92 에서 옵션은 다음과 같이 거래되었다.

Bought 10 OCT 192.5 CALL @ $6.30 (-)

Sold     10 OCT 197.5 CALL @ $4.10 (+)

                                      __________ 
                                         -$2.20

따라서, call position 에서 $-\$0.30$ 손해를 보았다. (10 계약 300 불). 그러나 more OTM 으로 adjust 할 생각이었으므로 다음과 같은 추가 position 을 구매한다.

Sold     10 OCT 200 CALL @ $3.00 (-)

Bought 10 OCT 205 CALL @ $1.80 (+)

                                   ____________ 
                                       +$1.20

이 결과로 $\$1.20$ (10 계약 1200 불) 의 credit 을 더 얻는다. 최종 position 은 다음과 같이 된다.

Bought 10 OCT 175 PUT    @ $2.55 (-) 

Sold     10 OCT 180 PUT    @ $3.80 (+) 

Sold     10 OCT 200 CALL @ $3.00 (+)

Bought 10 OCT 205 CALL @ $1.80 (-)

Maximum profit of $2.45 ($2450 for ten contracts)

Maximum loss of $2.55 ($2550 for ten contracts)

Our breakeven points at $\$177.55$ and $\$202.45$ which is the price range where XAU can move and we still dont lose money at expiration.

And remember that momentarily we have a loss of $300 for doing one adjustment. This loss will be substracted from our final profit in order to obtain our final real profit on this trade.

```python
ironCondor((175, 2.55), (180, 3.80), (200, 3.00), (205, 1.80),  (170, 210, 1))
```
