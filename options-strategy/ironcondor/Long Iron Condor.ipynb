# 콘도르 매수 - 성공확률 11%

    KOSPI 200 을 4 개의 다른 행사가로 2 개 매수, 2 개 매도
    
    290      1.68 풋 매도
    292.5   2.47 풋 매수
    297.5   4.93 풋 매수
    300      6.72 풋 매도

```python
st = np.arange(280, 310, 1)
total_payoff = np.zeros(st.shape)
fig, ax = plt.subplots()

strike1 = 290
premium = 1.68

payoff = short_put_payoff(st, strike1, premium)
plot_spread(st, payoff, strike1, fig, ax, 'Strike {} put short'.format(strike1),'dashed')
total_payoff += payoff

strike2 = 292.5
premium = 2.47

payoff = long_put_payoff(st, strike2, premium)
plot_spread(st, payoff, strike2, fig, ax, 'Strike {} put long'.format(strike2),'dashed')
total_payoff += payoff

strike3 = 297.5
premium = 4.93

payoff = long_put_payoff(st, strike3, premium)
plot_spread(st, payoff, strike3, fig, ax, 'Strike {} put long'.format(strike3),'dashed')
total_payoff += payoff

strike4 = 300
premium = 6.72

payoff = short_put_payoff(st, strike4, premium)
plot_spread(st, payoff, strike4, fig, ax, 'Strike {} put short'.format(strike4),'dashed')
total_payoff += payoff

plot_spread(st, total_payoff, strike4, fig, ax, 'Net Spread','solid')

plt.title('Condor 매수');

print('Max Profit = {:.2f}'.format(max(total_payoff)))
print('Max Loss = {:.2f}'.format(min(total_payoff)))
```
