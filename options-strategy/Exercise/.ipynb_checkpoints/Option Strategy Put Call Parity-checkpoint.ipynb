# 예제 8-1 Put Call Parity

A 기업의 주식에 대한 콜옵션과 풋옵션의 행사가격이 모두 33,000 으로 동일하며, 두 옵션은 유럽형이다. 만기는 1 년후. 현재의 주가는 28,000 이며 배당은 없다. 무위험 이자율은 10% 이다.

주식, 풋옵션, 콜옵션을 이용하

```python
# Straddle

total_premium = 0
multiplier = 500000
st = np.arange(170, 180, 1)
total_payoff = np.zeros(st.shape)
lower_strike = 175.0
higher_strike = 175.0

fig, ax = plt.subplots()

strike = higher_strike  # 높은 행사가의 콜옵션 1 개 매도
premium = 1.6

payoff = short_call_payoff(st, strike, premium, contracts=1)
plot_spread(st, payoff, strike, fig, ax, 'Strike {} call short'.format(strike),'dashed')
total_payoff += payoff
total_premium += premium

strike = lower_strike  # 낮은 행사가의 풋옵션 1 개 매도
premium = 1.0

payoff = short_put_payoff(st, strike, premium, contracts=1)
plot_spread(st, payoff, strike, fig, ax, 'Strike {} put short'.format(strike),'dashed')
total_payoff += payoff
total_premium += premium

plot_spread(st, total_payoff, strike, fig, ax, 'Net Spread','solid')

lower_even = lower_strike - total_premium
higher_even = higher_strike + total_premium
print('이익발생범위: {0:,.0f} < sT < {1:,.0f}'.format(
                lower_even*multiplier, higher_even*multiplier))
print('최대가능이익: {:,.0f}'.format(np.max(total_payoff)*multiplier))
```

```python
# Straddle

total_premium = 0
multiplier = 500000
st = np.arange(170, 180, 1)
total_payoff = np.zeros(st.shape)
lower_strike = 172.5
middle_strike = 175.0
higher_strike = 177.5

fig, ax = plt.subplots()

strike = middle_strike  # 높은 행사가의 콜옵션 2 개 매도
premium = 1.6
contracts = 2

payoff = short_call_payoff(st, strike, premium, contracts=contracts)
plot_spread(st, payoff, strike, fig, ax, 'Strike {} call short'.format(strike),'dashed')
total_payoff += payoff
total_premium += premium * contracts

strike = lower_strike  # 낮은 행사가의 콜옵션 1 개 매입
premium = 2.4
contracts = 1

payoff = long_call_payoff(st, strike, premium, contracts=contracts)
plot_spread(st, payoff, strike, fig, ax, 'Strike {} put long'.format(strike),'dashed')
total_payoff += payoff
total_premium -= premium

strike = higher_strike  # 높은 행사가의 콜옵션 1 개 매입
premium = 1.4
contracts = 1

payoff = long_call_payoff(st, strike, premium, contracts=contracts)
plot_spread(st, payoff, strike, fig, ax, 'Strike {} put long'.format(strike),'dashed')
total_payoff += payoff
total_premium -= premium

plot_spread(st, total_payoff, strike, fig, ax, 'Net Spread','solid')

lower_even = lower_strike - total_premium
higher_even = higher_strike + total_premium
print('이익발생범위: {0:,.0f} < sT < {1:,.0f}'.format(
                lower_even*multiplier, higher_even*multiplier))
print('최대가능이익: {:,.0f}'.format(np.max(total_payoff)*multiplier))
print('최대가능손실: {:,.0f}'.format(np.min(total_payoff)*multiplier))
```
