## 양매도

만기시 주가가 call 행사가보다 낮고 put 행사가보다 높아야 한다. 예측이 빗나가면 손실은 무한대가 된다.

```python
# short strangle
st = np.arange(290, 305, 1)
total_payoff = np.zeros(st.shape)
fig, ax = plt.subplots()

# 콜옵션 매도
strike1 = 300.0
premium = 1.29
contracts = 1
multiplier = 250000

payoff = short_call_payoff(st, strike1, premium, contracts, multiplier)
plot_spread(st, payoff, strike1, fig, ax, 
                'Strike {} call_long'.format(strike1),'--')
total_payoff += payoff

# 동일한 만기의 풋옵션 매도
strike2 = 295
premium = 1.17
contracts = 1
multiplier = 250000

payoff = short_put_payoff(st, strike2, premium, contracts, multiplier)
plot_spread(st, payoff, strike2, fig, ax, 
                'Strike {} put_long'.format(strike2),'--')
total_payoff += payoff

plot_spread(st, total_payoff, strike2, fig, ax, 'Net Spread','-')
plt.title("Short Strangle")

print('Max Profit = {:,.2f}'.format(max(total_payoff)))
print('Max Loss = {:,.2f}'.format(min(total_payoff)))
```
