# Iron Condor

```python
import os
import sys
sys.path.append(os.path.dirname(os.path.abspath(os.path.dirname('./'))))

from payoff import *
```

```python
def ironCondor(put_long, put_short, call_short, call_long, st_range):
    
    st = np.arange(st_range[0], st_range[1], st_range[2])
    total_payoff = np.zeros(st.shape)
    fig, ax = plt.subplots()
    net_premium = 0

    strike1, premium = put_long
    
    payoff = long_put_payoff(st, strike1, premium)
    plot_spread(st, payoff, strike1, fig, ax, 'Strike {} put_long'.format(strike1))
    total_payoff += payoff
    net_premium -= premium

    strike2, premium = put_short

    payoff = short_put_payoff(st, strike2, premium)
    plot_spread(st, payoff, strike2, fig, ax, 'Strike {} put_short'.format(strike2))
    total_payoff += payoff
    net_premium += premium

    strike3, premium = call_short

    payoff = short_call_payoff(st, strike3, premium)
    plot_spread(st, payoff, strike3, fig, ax, 'Strike {} call_short'.format(strike3))
    total_payoff += payoff
    net_premium += premium

    strike4, premium = call_long

    payoff = long_call_payoff(st, strike4, premium)
    plot_spread(st, payoff, strike4, fig, ax, 'Strike {} call_long'.format(strike4))
    total_payoff += payoff
    net_premium -= premium
    
    plot_spread(st, total_payoff, strike4, fig, ax, 'Net Spread')

    plt.title('Iron Condor');

    print('Max Profit = {:.2f}'.format(max(total_payoff)))
    print('Max Loss = {:.2f}'.format(min(total_payoff)))
    print('Net Premium = {:.2f}'.format(net_premium))
```

    Iron condor 는 상대적으로 낮은 payoff 를 갖지만 잠재적인 손실도 매우 제한적이다.
