# 예제 8-1 Put Call Parity

A 기업의 주식에 대한 콜옵션과 풋옵션의 행사가격이 모두 33,000 으로 동일하며, 두 옵션은 유럽형이다. 만기는 1 년후. 현재의 주가는 28,000 이며 배당은 없다. 무위험 이자율은 10% 이다.

주식, 풋옵션, 콜옵션을 이용한 무위험 포트폴리오를 구성하라.

콜옵션의 현재 가격이 4,000 원일 경우 풋옵션의 현재 가치는 얼마인가 ?

```python
total_premium = 0
multiplier = 1
st = np.arange(30000, 36000, 1)
total_payoff = np.zeros(st.shape)
strike = 33000

fig, ax = plt.subplots()

# 주식 매입
plt.plot(st,st-strike, '--', label='stock')
total_payoff += st-strike

# 풋옵션 매입
premium = 0

payoff = long_put_payoff(st, strike, premium, contracts=1)
plot_spread(st, payoff, strike, fig, ax, 'Strike {} put long'.format(strike),'dashed')
total_payoff += payoff

# 콜옵션 매도
payoff = short_call_payoff(st, strike, premium, contracts=1)
plot_spread(st, payoff, strike, fig, ax, 'Strike {} call short'.format(strike),'dashed')
total_payoff += payoff

plot_spread(st, total_payoff, strike, fig, ax, 'Net Spread','solid')

lower_even = strike - total_premium
higher_even = strike + total_premium
print('이익발생범위: {0:,.0f} < sT < {1:,.0f}'.format(
                lower_even*multiplier, higher_even*multiplier))
print('최대가능이익: {:,.0f}'.format(np.max(total_payoff)*multiplier))
```

P+S-C = X / 1+Rt

```python
33000 / (1+0.1) + 4000 - 28000
```
