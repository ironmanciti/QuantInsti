```python
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
import numpy as np
import seaborn as sns
sns.set()
%matplotlib inline
```

```python
plt.plot([1.0, 2.0, 3.0, 4.0],[1.0, 2.0, 3.0, 4.0], color='lightblue', linewidth=5)
plt.scatter(np.linspace(0,5,4),[0.5, 0.8, 3.4, 4.5], color='darkgreen', marker='^')
plt.xlim(-2.0,5.0)
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(221)
ax1.plot([1.0, 2.0, 3.0, 4.0],[1.0, 2.0, 3.0, 4.0], linewidth='4', color='darkblue')
ax1.scatter([1.0, 2.0, 3.0, 4.0],[0.5, 0.8, 3.4, 4.5], marker='o')
ax1.set_xlim(-1, 4.5)
```

```python
fig = plt.figure(figsize=(20,10))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)
ax1.bar([0,3,5],[3,5,8])
ax1.set_title('TEST BAR')
ax2.barh([0, 3, 5],[1,4,6])
ax2.set_title('TEST BARH')
plt.suptitle('TEST BAR & BARH')
```

```python
fig = plt.figure(figsize=(20,10))
ax1 = fig.add_subplot(1,3,1)
ax1.bar(np.linspace(0,3,3),[3,4,7])
ax2 = fig.add_subplot(1,3,2)
ax2.barh([1,2,3],[3,4,5])
ax2.axhline(2.0, color='red')
ax3 = fig.add_subplot(1,3,3)
ax3.scatter(np.linspace(0,10,5), np.linspace(3,5,5))
ax3.axvline(4.0)
fig.delaxes(ax2)   #delete axes
fig.add_axes(ax2)  # restore ax2
plt.savefig("foo.png", transparent=True)
pp = PdfPages("pdf.pdf")
pp.savefig()
pp.close()
```
